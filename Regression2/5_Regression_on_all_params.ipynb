## Линейная функция от m параметров

$$y = \theta_0 + \theta_1 x_1 + \theta_2 x_2 + \theta_3 x_3 + \ldots + \theta_m x_m$$

Представим строку с данными как:

${X}_i = \begin{pmatrix}1 & x_{i, 1} &  x_{i, 2} & x_{i, 3} & \ldots, x_{i, m}\end{pmatrix}.$

В данном случае $m = 13$.

Представим параметры линейной функции 
$\Theta = \begin{pmatrix}\theta_0& \theta_1& \theta_2& \theta_3& \ldots, \theta_{m}\end{pmatrix}^T$

Тогда линейную функцию от N аргументов можно представить в векторном виде:

\begin{equation*}
\tilde{y}_i= \begin{pmatrix}x_{i, 0} & x_{i, 1}& x_{i, 2}& \ldots& x_{i, m}\end{pmatrix} \begin{pmatrix}
    \theta_0\\
    \theta_1\\
    \theta_2\\
    \cdots\\
    \theta_{m}\\
  \end{pmatrix} =  \sum_{j=0}^{m} \theta_j x_{i, j} = X_i\Theta.
\end{equation*}

 $x_{i, 0} = 1$

И для предсказания всех занчений 

$$y = X\Theta$$

## Реализации линейной функция 

```python
def linear_function(X, Theta):
    return np.dot(X, Theta)
```

## Функция ошибки

Определим функцию ошибки от параметров $\Theta$ :

\begin{equation*}
Loss(\Theta) = \dfrac{1}{N}\sum_{i=1}^{N}{(\tilde{y_i} - y_i)^2}= \dfrac{1}{N} \sum_{i=1}^{N}{(X_i\Theta - y_i)^2}
\end{equation*}

Где $N$ - это количество квартир, $y_i$ - реальная цена квартиры, $\tilde{y_i}$ - предсказанная цена для $i$-oй квартиры.   

## Реализации функции ошибки

\begin{equation*}
Loss(\Theta) = \dfrac{1}{N}\sum_{i=1}^{N}{(\tilde{y_i} - y_i)^2}= \dfrac{1}{N} \sum_{i=1}^{N}{(X_i\Theta - y_i)^2}
\end{equation*}

## Градиент функции ошибки

Частная производная $j$-го параметра:

\begin{equation*}
\dfrac{\partial Loss(\Theta)}{\partial \theta_j} = \dfrac{2}{N} \sum_{i=1}^{N}(X_i\Theta - y_i) x_{ij}
\end{equation*}

Тогда градиент функции ошибки линейной функции от $m$ параметров:

\begin{equation*}
\nabla Loss(\Theta) =
   \begin{bmatrix}
   \dfrac{2}{N} \sum_{i=1}^{N} (X_i\Theta - y_i)x_{i,0}\\
   \cdots \\
   \dfrac{2}{N} \sum_{i=1}^{N} (X_i\Theta - y_i)x_{i,j}\\
   \cdots \\
   \dfrac{2}{N} \sum_{i=1}^{N} (X_i\Theta - y_i)x_{i,m}\\
   \end{bmatrix}
\end{equation*}

## Реализация подсчета градиента

```python
def gradient_function(Theta, X, y_true):
    grad = np.zeros_like(Theta)
    y_pred = linear_function(X, Theta)
    for j in range(Theta.shape[0]):       
        grad[j] = 2*np.mean((y_pred - y_true)* X[:,j])
    return grad
```

## Градиентный спуск

* Выбираем случайное значение для $\Theta = (\theta_0, \theta_1, \ldots, \theta_m)^T $
* Повторить $iter$ раз:

    &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp; $\Theta_{new} = \Theta - \alpha \nabla Loss(\Theta)$

Где $\alpha$ это коэффициент, который мы выбираем.

## Реализация градиентного спуска

```python
def gradient_descent(theta_init, X, y, alpha, iters):        
    theta = theta_init.copy()
    for i in range (iters):
        theta = theta - alpha * gradient_function(theta, X, y)

    return theta
```
