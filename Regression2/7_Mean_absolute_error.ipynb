# Mean absolute error

В качестве функции ошибки для обучения линейной регрессии до этого всегда использовалась MSE (Mean squred error):
    
$MSE = \frac{1}{N}\sum_{i=0}^{N}{(\hat{y_i} - y_i)^2}= \frac{1}{N} \sum_{i=0}^{N}{(X_i\Theta - y_i)^2}$
    

MSE не единственная функция ошибки, которая применяется для обучения регрессии.

На ряду с MSE также применяется MAE (Mean absolute error):

$MAE = \frac{1}{N}\sum_{i=0}^{N}{|\hat{y_i} - y_i|}= \frac{1}{N} \sum_{i=0}^{N}{|X_i\Theta - y_i|}$

Это то, насколько в среднем ошибается модель.

## Производная МАЕ

В общем случае производная от $|x|$ не определена в точке 0, во всех остальных случаях ее можно определить, как $|x|'= \dfrac{|x|}{x}$.

При $x > 0, \ |x| = x, \ \dfrac{|x|}{x} = \dfrac{x}{x} = 1$.

При $x < 0, \ |x| = -x, \ \dfrac{|x|}{x} = \dfrac{-x}{x} = -1$.

В нашем случае мы можем доопределить производную от $|x|$ в нуле значением 0. Тогда она совпадет с функцией знака:
\begin{equation*}
 sign(x) = 
 \begin{cases}
   1 &\text{x > 0}\\
   0 &\text{x = 0}\\
   -1 &\text{x < 0}
 \end{cases}
\end{equation*}

Производная от $|f(x)|$ не определена в точке $x_0$, где $f(x_0) = 0$ во всех остальных случаях ее можно определить, как $|f(x)|'= \dfrac{|f(x)|}{f(x)}f'(x) = sign(f(x))f'(x)$ .

Теперь мы можем посчитать градиент функции потерь:  
\begin{equation*}
\frac{\partial Loss(\Theta)}{\partial \theta_j} = \frac{1}{N} \sum_{i=1}^{N} sign(X_i\Theta - y_i) x_{ij}
\end{equation*}

\begin{equation*}
\nabla Loss(\Theta) = 
 \begin{bmatrix}
   \frac{1}{N} \sum_{i=1}^{N} sign(X_i\Theta - y_i)x_{i0}\\
   \cdots\\
   \frac{1}{N} \sum_{i=1}^{N} sign(X_i\Theta - y_i) x_{ij} \\
   \cdots \\
   \frac{1}{N} \sum_{i=1}^{N} sign(X_i\Theta - y_i)x_{im}\\
 \end{bmatrix}
\end{equation*}
