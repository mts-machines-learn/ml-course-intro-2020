## Функция от двух параметров

\begin{equation*}
\tilde{y} = f(x_1, x_2) = \theta_0 + \theta_1 x_1 + \theta_2 x_2.
\end{equation*}

* $\tilde{y}$ - предсказанная цена.

* $x_1$ - Cреднее количество комнат.  

* $x_2$ - LSTAT.

* $\theta_0, \theta_1, \theta_2$ - коэффициенты функции.

Визуализируем как выглядит эта функция.
